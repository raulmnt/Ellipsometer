```python
thickness=25
d_list = [inf,thickness,inf] #in nm
lambda_list = linspace(300,800,51) #in nm
Si_nk_fn = interp1d(Si_nk_data[:,0].real,
                          Si_nk_data[:,1], kind='quadratic')
SiO2_nk_fn = interp1d(SiO2_nk_data[:,0].real,
                          SiO2_nk_data[:,1], kind='quadratic')
# calculate psi and delta for first angle
psi_1=[]
delta_1=[]

for lambda_vac in lambda_list:
            n_list = [1,SiO2_nk_fn(lambda_vac),Si_nk_fn(lambda_vac)]
            e_data=ellips(n_list, d_list, 65*degree, lambda_vac) #in nm
            psi_1.append(e_data['psi']/degree) # angle in degrees
            delta_1.append((pi-e_data['Delta'])/degree) # angle in degrees


#  calculate psi and delta for second degrees
psi_2=[]
delta_2=[]

for lambda_vac in lambda_list:
            n_list = [1,SiO2_nk_fn(lambda_vac),Si_nk_fn(lambda_vac)]
            e_data=ellips(n_list, d_list, 75*degree, lambda_vac) #in nm
            psi_2.append(e_data['psi']/degree) # angle in degrees
            delta_2.append((pi-e_data['Delta'])/degree) # angle in degrees       

```

```python
# plot modeled and experimental delta
plt.figure()
plt.plot(lambda_list,delta_1, raw_data_1.lambda_vac ,raw_data_1.delta_exp,
         lambda_list,delta_2, raw_data_2.lambda_vac , raw_data_2.delta_exp)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Delta (degrees)')
plt.title('Ellipsometric parameters for first angle');

# plot modeled and experimental psi
plt.figure()
plt.plot(lambda_list , psi_1 , raw_data_1.lambda_vac , raw_data_1.psi_exp,
        lambda_list , psi_2 , raw_data_2.lambda_vac , raw_data_2.psi_exp)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Psi (degrees)')
plt.title('Ellipsometric parameters for second angle');

# make a dataframe with the values from the model
mod=pd.DataFrame({"psi_1":psi_1,
                   "delta_1":delta_1,"psi_2":psi_2,
                   "delta_2":delta_2})

# calculate error considering both angles and all wavelengths
xi_1=((mod['psi_1'] - raw_data_1['psi_exp'])/(raw_data_1['psi_exp_sigma']))**2+((mod['delta_1'] - raw_data_1['delta_exp'])/(raw_data_1['delta_exp_sigma']))**2
xi_2=((mod['psi_2'] - raw_data_2['psi_exp'])/(raw_data_2['psi_exp_sigma']))**2+((mod['delta_2'] - raw_data_2['delta_exp'])/(raw_data_2['delta_exp_sigma']))**2
MSE=((xi_1.sum(axis=0)+xi_2.sum(axis=0))/(2*(len(xi_1)+len(xi_2))-1))**(1/2)
print('MSE=', MSE)
```
